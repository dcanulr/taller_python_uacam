```python
equis = np.linspace(0, 2*np.pi, 50)
seno = np.sin(equis)
coseno = np.cos(equis)
tangente = np.tan(equis)
```

```python
fig, ax = plt.subplots(figsize=(15,5))

ax.plot(equis, seno, label="Seno", marker="d", ls="", color="black")
ax.plot(equis, coseno, label="Coseno", color="red", ls="--")
ax.set_xlabel("Radianes")
ax.set_ylabel("Valor de la función")
ax.set_title("Mi primera gráfica")
ax.set_xticks([0, 0.5*np.pi, np.pi, 3/2*np.pi, 2*np.pi], 
              ["0", "1/2$\pi$","$\pi$", "$3/2\pi$", "2$\pi$"])
ax.grid(color="k", ls="--", alpha=0.4)
# ax.set_ylim(-5,5)"
ax.legend()

plt.savefig("grafica1.png")
```
