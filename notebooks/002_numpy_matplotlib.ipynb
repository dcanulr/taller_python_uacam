```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
equis = np.linspace(0, 2*np.pi)
equis
```

```python
np.sin(2)
```

```python
np.sin(np.pi)
```

```python
seno = np.sin(equis)
seno
```

```python
coseno = np.cos(equis)
tangente = np.tan(equis)
```

![](../figs/anatomy_figure.png)

```python
fig, ax = plt.subplots()
ax.plot(equis, seno)
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(equis, seno, label="Seno")
ax.set_xlabel("El eje x")
ax.set_ylabel("El eje y")
ax.set_title("Mi primera gráfica")
ax.legend()

# plt.savefig("mi_primer_grafica.png")
plt.savefig("mi_primer_grafica.pdf")

#Modificar y guardar
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(equis, seno, label="Seno")
ax.plot(equis, coseno, label="Coseno")
ax.set_xlabel("El eje x")
ax.set_ylabel("El eje y")
ax.set_title("Mi primera gráfica")
ax.legend()

# plt.savefig("mi_primer_grafica.png")
# plt.savefig("mi_primer_grafica.pdf")
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(equis, seno, label="Seno")
ax.plot(equis, coseno, label="Coseno")
ax.set_xlabel("El eje x")
ax.set_ylabel("El eje y")
ax.set_title("Mi primera gráfica")
ax.set_xticks([0, np.pi/2, np.pi, 1.5*np.pi, 2*np.pi], ["0", "$\pi/2$","$\pi$","$3/2\pi$", "$2\pi$"])
ax.legend()
```

```python
fig, ax = plt.subplots(nrows=3)
ax[0].plot(equis, seno)
ax[1].plot(equis, coseno)
ax[2].plot(equis, tangente)
plt.tight_layout()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax[0,0].plot(equis, seno)
ax[1,1].plot(equis, coseno)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax = ax.flatten()
ax[0].plot(equis, seno)
ax[3].plot(equis, coseno)
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax = ax.flatten()
ax[0].plot(equis, seno, color="red", ls="", marker="d", markersize=1)

ax[1].plot(equis, coseno, color="blue", ls="--", marker="3")

ax[2].plot(equis, tangente, color="green", marker="o", markersize=1)

ax[3].plot(equis, seno)
ax[3].plot(equis, coseno)
ax[3].plot(equis, tangente)
ax[3].set_ylim(-5,5)
ax[3].vlines(np.pi/2, -5, 5, ls="--", color="k")
ax[3].vlines(np.pi*3/2, -5, 5, ls="--", color="k")
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, sharex="col")
ax = ax.flatten()
ax[0].plot(equis, seno, color="red", ls="", marker="d", markersize=1)

ax[1].plot(equis, coseno, color="blue", ls="--", marker="3")
ax[1].sharey(ax[0])

ax[2].plot(equis, tangente, color="green", marker="o", markersize=1)
ax[2].set_xticks([0, np.pi/2, np.pi, 1.5*np.pi, 2*np.pi], ["0", "$\pi/2$","$\pi$","$3/2\pi$", "$2\pi$"])

ax[3].plot(equis, seno, label="Seno")
ax[3].plot(equis, coseno, label="Coseno")
ax[3].plot(equis, tangente, label="Tangente")
ax[3].set_xticks([0, np.pi/2, np.pi, 1.5*np.pi, 2*np.pi], ["0", "$\pi/2$","$\pi$","$3/2\pi$", "$2\pi$"])
ax[3].set_ylim(-5,5)
ax[3].vlines(np.pi/2, -5, 5, ls="--", color="k")
ax[3].vlines(np.pi*3/2, -5, 5, ls="--", color="k")
ax[3].legend()
```
