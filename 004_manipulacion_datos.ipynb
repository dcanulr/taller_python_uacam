```python
diciembre["Suma"] = diciembre.Temp + diciembre.HR
```

```python
diciembre
```

```python
def sensacionT(temp, vel):
    st = 13.1267+0.6215*temp-11.37*vel**0.16+0.3965*temp*vel**0.16
    return st
```

```python
sensacionT(35, 80)
```

```python
diciembre["SensacionT"] = sensacionT(diciembre.Temp, diciembre.WS)
```

```python
fig, ax = plt.subplots(figsize=(10,4))
ax.plot(diciembre.Temp, label="Temperatura")
ax.plot(diciembre.SensacionT, label="Sensacion TÃ©rmica")
ax.plot(diciembre.HR, label="Humedad Relativa")
ax.legend()
```
